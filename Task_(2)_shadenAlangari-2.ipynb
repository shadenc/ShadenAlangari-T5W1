## 1. Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/datasets/Task (2) Dataset.csv')
```

```python
df.head()
```

```python
df.sample(5)
```

### Perform summary statistics on the dataset.

```python
df.describe()
```

### Identify and analyze patterns in the data.

```python
# Select numerical variables
numerical_df = df.select_dtypes(include=['number'])
# Perform correlation
correlation_matrix = numerical_df.corr()
correlation_matrix
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['TrafficIndexLive'], kde=True)
plt.title('Distribution of Traffic Index Live')
plt.xlabel('Traffic Index Live')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
# Plotting the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

## 2. Data Visualization

* Ensure the visualizations are clear and informative.

### Create visualizations to illustrate the findings from the EDA.


```python
# prompt: ### Create visualizations to illustrate the findings from the EDA.

# Example 2: Scatter plot of Traffic Index vs. Jams Count
plt.figure(figsize=(8, 6))
# Check the correct column name in your DataFrame. It might be 'JamsCount' instead of 'Jamscount'.
sns.scatterplot(x='JamsCount', y='TrafficIndexLive', data=df)  # Changed 'Jamscount' to 'JamsCount'
plt.title('Traffic Index vs. Jams Count')
plt.xlabel('Jams Count')
plt.ylabel('Traffic Index Live')
plt.grid(True)
plt.show()
```

```python
median_traffic_index = df['TrafficIndexLive'].median()
above_median = (df['TrafficIndexLive'] > median_traffic_index).sum()
below_median = (df['TrafficIndexLive'] <= median_traffic_index).sum()
plt.figure(figsize=(8, 8))
plt.pie([above_median, below_median], labels=['Above Median', 'Below Median'], autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Traffic Index Live Above and Below Median')
plt.show()
```

### Use appropriate plots such as histograms, bar charts, pie charts, scatter plots, and heatmaps.

```python
# Group the data by city and calculate the average traffic index
city_traffic = df.groupby('City')['TrafficIndexLive'].mean().sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(12, 6))
city_traffic.plot(kind='bar', color='skyblue')
plt.title('Average Traffic Index by City')
plt.xlabel('City')
plt.ylabel('Average Traffic Index')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
